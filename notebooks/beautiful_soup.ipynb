## Get names of munros from A-Z page

```python
def make_soup(url):
    '''
    '''

    page = requests.get(url)
    
    return BeautifulSoup(page.content, "html.parser")
```

```python
url_base = "https://www.walkhighlands.co.uk"
URL = url_base + "/munros/munros-a-z"

soup = make_soup(URL)

# NB prettify makes bs4 object into a str object
print(soup.prettify())
```

```python
results = soup.find_all(class_="table1")
print(type(soup))
print(len(results))
```

```python
# This one is pretty neat but only gets half of them
results6 = soup.table.find_all('a')
print(results6)
print(len(results6))
```

```python
import numpy as np

results5 = soup.select('.table1')
print(len(results5))

# Loop over table elements
results8 = []
for i in range(len(results5)):
    results3 = [tag.get('href') for tag in results5[i].find_all('a')][3:]
    results8.append(results3)

results8 = list(np.concatenate(results8).flat)

print(results8)
#print(len(results8))
```

## Find walk link in munro page

```python
munro1_url = "https://www.walkhighlands.co.uk/munros/" + results8[0]
print(munro1_url)
```

```python
munro_page = requests.get(munro1_url)

munro_soup = BeautifulSoup(munro_page.content, "html.parser")
#print(munro_soup)
```

```python
results9 = munro_soup.h2.next_sibling.next_sibling.find('a').get('href')
print(results9)
```

```python
walk_url = url_base + results9
print(walk_url)
```

## Get start point coords

```python
soup2 = make_soup(walk_url)
print(soup2)
```

```python
results10 = soup2.find_all('script')[2].text.split(';')[0:2]

print(results10)
print(type(results10))

coords = [float(element[9:]) for element in results10]
print(coords, type(coords[0]))
```

```python
def get_start_point_coords_of_walk(soup):
    '''
    returns a list of lat, lon start point coordinates as floats
    '''

    coords = soup.find_all('script', type='text/javascript')[1].text.split(';')[0:2]

    ''' testing block
    coords = soup.find_all('script', type='text/javascript')[1]
    print(coords)
    coords = coords.text
    print(coords)
    coords = coords.split(';')[0:2]
    '''
    
    #print('start coords:', coords)
    return [float(element[9:]) for element in coords]
```

## Do for all munros

```python
import pandas as pd
# list of all munros results8

# convert munros to dataframe
munros_df = pd.DataFrame(results8, columns = ['munro'])

# make empty lists to store lat and lon start points
lat = []
lon = []

# loop over list
for munro in results8:
    # go to munro page
    munro_url = (url_base + "/munros/" + munro).strip()
    #print(munro_url)
    munro_soup = make_soup(munro_url)
    # go to walk page
    walk_href = munro_soup.h2.next_sibling.next_sibling.find('a').get('href')
    
    walk_url = (url_base + walk_href).strip()
    #print(repr(walk_url))
    # get start point coordinates
    walk_soup = make_soup(walk_url)
    #print('SOUP', walk_soup)
    start_coords = get_start_point_coords_of_walk(walk_soup)
    
    # append lat, lon to lists
    lat.append(start_coords[0])
    lon.append(start_coords[1])

#   put them in dataframe with name of munro

print('lat:', lat, 'lon:', lon)
munros_df['lat'] = lat
munros_df['lon'] = lon

print(munros_df)

```
