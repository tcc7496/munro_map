## Get names of munros from A-Z page

```python
URL = "https://www.walkhighlands.co.uk/munros/munros-a-z"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
results = soup.find_all(class_="table1")
print(type(soup))
print(len(results))
```

```python
# This one is pretty neat but only gets half of them
results6 = soup.table.find_all('a')
print(results6)
print(len(results6))
```

```python
import numpy as np

results5 = soup.select('.table1')
print(len(results5))

# Loop over table elements
results8 = []
for i in range(len(results5)):
    results3 = [tag.get('href') for tag in results5[i].find_all('a')][3:]
    results8.append(results3)

results8 = list(np.concatenate(results8).flat)

#print(results8)
#print(len(results8))
```
